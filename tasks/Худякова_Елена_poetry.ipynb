Пакетный менеджер. В современном мире большинство языков программирования позволяют создавать независимые компоненты, которые можно использовать в разных проектах одновременно. Разработчик может соединить эти компоненты в единый пакет, который могут использовать другие разработчики. И для их скачивания, добавления и установки в свой пакет используются пакетные менеджеры. 

Большинство пакетных менеджеров используют специальный файл, в котором можно указать от каких пакетов зависит используемый код. Также эти пакеты могут зависеть от других. Тем самым, получается сложная структура, которую необходимо распределять по полочкам. Задача пакетного менеджера – разрешать эти зависимости. Поэтому иногда их называют менеджерами зависимостей. 
Все началось с того, что создателю Poetry Себастьену Юстасу потребовался единый инструмент для управления проектами от начала до конца, надежный и интуитивно понятный, который бы мог использоваться и в рамках сообщества. Одного лишь менеджера зависимостей было недостаточно, чтобы управлять запуском тестов, процессом развертывания и всем созависимым окружением. Этот функционал находится за гранью возможностей обычных пакетных менеджеров, таких как Pip или Conda. Так появился Python Poetry.

•	Conda – используется в Data Science, но не особо популярен среди разработчиков.
•	Pip – среди разработчиков чаще используется именно он, однако, существуют хорошие альтернативы. Именно к его альтернативам относится Poetry.

Рассмотрим подробнее Poetry – это инструмент для управления зависимостями и сборкой пакетов в Python. В Poetry представлен полный набор инструментов, которые могут понадобиться для детерминированного управления проектами на Python. В том числе, сборка пакетов, поддержка разных версий языка, тестирование и развертывание проектов. Ключевой особенностью Poetry Python является полное управление библиотеками, заявленными для вашего проекта, включая их установку, обновление и публикацию. Функционал командной строки расширяется с помощью плагинов, а предустановленные значения при необходимости легко заменяются на свои варианты. В целом, продукт имеет дружественный интерфейс.
 Для выполнения своих задач Poetry нужен только один файл: новый, стандартизированный pyproject.toml. Другими словами, Poetry использует pyproject.toml для замены setup.py, requirements.txt, setup.cfg, MANIFEST.in и Pipfile.
Poetry также определит, находитесь ли вы внутри виртуальной среды и установит пакеты соответственно. Таким образом, инструмент может быть установлен глобально и использоваться везде. Poetry также поставляется с полноценной библиотекой разрешения зависимостей.

Упаковочные системы и управление зависимостями в Python довольно запутаны и трудны для понимания новичками. Даже для опытных разработчиков иногда может быть громоздко создавать все файлы, необходимые в проекте. Поэтому на помощь приходит инструмент, который ограничивает все действия одним файлом конфигурации: управление зависимостями, упаковка и публикация.

### *Возможности Poetry Python*
•	Создавать и упаковывать проекты с помощью одной команды.
•	Публиковать проекты в PyPI и приватных репозиториях.
•	Проверять состояния зависимостей одной командой.
•	Фиксировать версии зависимостей.
•	Использовать ваши настроенные виртуальные сервера или создавать собственное виртуальное окружение для полной изоляции от вашей системы.

### *Достоинства*
Poetry четко задает формат пакета в декларативном виде, в формате файла конфигурации. Таким образом, файлы легче систематизировать и собирать другими программами и использовать другим приложениями, ведь данные стандартизированные. При создании нового пакета создается файл с метаданными, который описывает все зависимости созданного пакета. Все пакетные менеджеры для питона умеют читать этот файл с метаданными и разрешать зависимости, используя его. Таким образом, установка производится быстрее и требует меньше трафика. Именно умением создавать пакеты Poetry отличается от Pip. Отметим, что Poetry умеет разрешать сложные графы зависимости, которые не под силу другим пакетным менеджерам.

### *Недостатки*
Poetry – это очень молодой проект, иногда встречаются ошибки, из-за которых необходимо искать обходные пути, однако, найти их можно. Также инструмент имеет не такое большое сообщество пользователей, по сравнению с Pip. Не все сторонние программы поддерживают формат файлов Poetry, но это меняется со временем. 

# *Установка*
### *Установка на OSX и Linux*

Особенность скачивания и установки пакетов заключается в том, что они выполняются параллельно друг другу, отсюда выигрыш во времени. 
Этапы установки на Mac и Linux идентичны:
•	Проверить версию Python
(в консоли python --version, и, если версия Python по умолчанию 2.7, проверить python3 --version).
•	Чтобы установить актуальную версию вводим в консоль:




```python
curl -sSL https://raw.githubusercontent.com/python-poetry/poetry/master/get-poetry.py | python -
```

Если версия языка не устраивает или если для установки хочется воспользоваться версией питона 3.x, а по умолчанию стоит 2.7, используйте команду:



```python
curl -sSL https://raw.githubusercontent.com/python-poetry/poetry/master/get-poetry.py | python3 -
```

Обратите внимание: процесс установки может занять довольно длительное время.
Каталог bin, в который происходит установка инструмента, в Unix расположен в 
$HOME/.poetry/bin.

### *Установка на Windows*

Для инсталляции Poetry:
•	Откройте Powershell (Win + X далее i, или найдите PoweShell через поиск).
•	Введите команду:




```python
(Invoke-WebRequest -Uri https://raw.githubusercontent.com/python-poetry/poetry/master/get-poetry.py -UseBasicParsing).Content | python -
```

Перезапустите оболочку и все заработает. Папку установки для Windows можно найти в %USERPROFILE%\.poetry\bin.
Для корректной работы установщик должен иметь возможность найти исполняемые файлы py.exe -3 или py.exe -2.

### *Установка через PIP и PIPX*

Для данного варианта установки необходимо иметь на борту менеджер пакетов PIP (или PIPX). Для установки Poetry через PIP выполните команду:




```python
pip install --user poetry
```

Имейте в виду, что зависимости пакета Poetry могут вызвать конфликт с уже установленными пакетами в вашей системе.
Команды Pipx для установки, обновления и удаления инструмента:


```python
pipx install poetry
```

```python
pipx upgrade poetry
```

```python
pipx uninstall poetry
```

Pipx поддерживает только версии Python старше 3.6.
Корректность установки проверяется с помощью команды "poetry --version".

### *Poetry и виртуальное окружение*

По умолчанию, Poetry автоматически создает виртуальное окружение для вашего проекта в cache-директории:

•	для Windows – C:\Users\<username>\AppData\Local\pypoetry\Cache\virtualenvs;
•	для Linux – ~/.cache/pypoetry/virtualenvs;
•	для macOS – ~/Library/Caches/pypoetry/virtualenvs.

В эту директорию будут сохраняться все зависимости для проекта.



### *Конфигурация Poetry* 

Конфигурация Poetry настраивается с помощью команды poetry config. После первого запуска команды, Poetry создаст файл конфигурации config.toml, в котором будут храниться все заданные параметры. Находится файл в директории:

•	для Windows – C:\Users\<username>\AppData\Roaming\pypoetry;
•	для Linux – ~/.config/pypoetry;
•	для macOS – ~/Library/Application Support/pypoetry.

Для примера попробуем настроить Poetry так, чтобы он создавал виртуальное окружение virtualenvs не в cache-директории, а в корневом каталоге проекта. Для этого установим переменной virtualenvs.in-project значение true:



```python
poetry config virtualenvs.in-project true
```

Чтобы убедиться, что значение успешно установилась, выполним команду:


```python
poetry config virtualenvs.in-project
```

Чтобы посмотреть все текущие параметры Poetry, используйте команду poetry config --list. Теперь все зависимости будут устанавливаться в папку .venv в корне вашего проекта. Если параметр нужно удалить, используйте флаг --unset:

```python
poetry config virtualenvs.in-project --unset
```

# *Работа с Poetry*
### *Создание проекта*
Создаем пробный проект и присваиваем ему название "project-poetry":


```python
poetry new project-poetry
```

Произойдет автоматическое создание папки проекта со следующей структурой:

```python
project-poetry
    project-poetry
        __init__.py
    tests
        __init__.py
        test_project_poetry.py
    pyproject.toml
    README.rst
```

Сообщение "Created package first in first" здесь означает, что в проекте "first" был создан одноимённый пакет. Любой poetry-проект всегда содержит хотя бы один пакет.

Кроме пакета first, в проекте уже есть пакет tests с первым тестом. В файле README.rst содержится описание проекта. Расширение .rst говорит о том, что это файл формата reStructuredText, именно в данном формате и документируют код проектов Python. Это облегчённый язык разметки. Хорошо применим для создания простых веб-страниц и других документов, а также в качестве вспомогательного языка при написании комментариев в программном коде. Используется в системе подготовки технической документации Docutils.

И, наконец, самый важный файл в poetry-проекте — pyproject.toml. Это файл конфигурации проекта, описывающий всё, что нужно poetry знать, чтобы

•	управлять зависимостями проекта

•	запускать код на исполнение

•	запускать инструменты для разработки

•	собирать дистрибутив и публиковать его на PyPI

Отдельно стоит отметить последний пункт: poetry приучает к тому, чтобы вы всегда думали о проекте так, как будто вы будете его публиковать — любой "уважающий себя" пакет имеет хорошее описание и правильную структуру, ему всегда назначена какая-то версия. Стоит думать о таких вещах, даже если вы не собираетесь публиковать каждый ваш проект.

Расширение .toml говорит о том, что это файл формата TOML. Это формат конфигурационных файлов, спроектированный для обеспечения человекочитаемости с одной стороны и однозначного преобразования в ассоциативный массив с другой. Спецификация языка открыта и дополняется сообществом. Название «TOML» является акронимом «Tom’s Obvious, Minimal Language» (Очевидный язык Тома), имея в виду своего создателя, Tom Preston-Werner. TOML используется в ряде программных продуктов (проекты перечислены по ссылкам [8],[9]), его поддержка реализована в большом количестве языков программирования [10].

В файле pyproject.toml содержится описание Poetry-проекта – название, описание, используемый репозиторий, зависимости проекта и т.д. С помощью него легко организовать зависимости проекта (pyproject.toml можно назвать преемником устаревшего requirements.txt).

Если вы знакомы с Pipenv, то наверняка заметите, что pyproject.toml очень похож на Pipfile. Pipenv – менеджер зависимостей для Python-проектов. С его помощью можно создавать виртуальные среды и управлять зависимостями приложений. Pipenv решает ряд проблем, которые возникали при использовании pip, virtualenv и requirements.txt. Данный инструмент аналогичен npm у NodeJS, или composer у PHP, и является официально рекомендуемым менеджером зависимостей для Python.

После инициализации проекта содержание файла будет следующим:
[tool.poetry]

name = "project-poetry"

version = "0.1.0"

description = ""

authors = ["Alexander <info@pythonchik.ru>"]

[tool.poetry.dependencies]

python = "^3.8"

[tool.poetry.dev-dependencies]

pytest = "^5.2"

[build-system]

requires = ["poetry-core>=1.0.0"]

build-backend = "poetry.core.masonry.api"

Здесь строчки вида [tool.poetry] описывают секции, в секциях же находятся пары "ключ = значение". Важно помнить, что большинство ключей нельзя помещать "не в те" секции. Вы всегда можете посмотреть в документации, какие ключи бывают и каким секциям они принадлежат. Обратимся к каждой секции:

•	[tool.poetry] – содержит метаданные проекта, которые не относятся к пакетам зависимостей – название, описание, указание авторства и контактную информацию;

•	[tool.poetry.dependencies] – раздел содержит зависимости, необходимые для проекта;

•	[tool.poetry.dev-dependencies] – зависимости разработки (не войдут в сборку проекта, но необходимые для других аспектов – тестирования, документации и т. д.);

•	[build-system] – если вы используете Poetry для управления своим проектом, вы должны ссылаться на него в этом разделе (команды new и init автоматически добавляют информацию в раздел build-system).
Все переменные могут задаваться вручную в соответствующие места Pyproject.toml, а также с помощью команды add или добавляться автоматически средствами Poetry.

### *Сборка и настройка пакета*
Чтобы собрать пакет, запускаем команду build.


```python
poetry build
```

В результате в папке dist будут сформированы пакеты.
   
*Рисунок 1 - Сформированные пакеты в папке dist*

Если необходимо ограничить сборку определенным типом, можно просто использовать параметр -F.

```python
poetry build -F wheel # or sdist
```

Чтобы развернуть пакет, нужно настроить репозитории и учетные данные. Для этого применяется команда config.

```python
poetry config repositories.priv https://my.repository
```

В примере происходит настройка репозитория с именем priv и Url-адресом https://my.repository. Сохранить учетные данные для этого репозитория также можно с помощью команды config:

```python
poetry config http-basic.priv user password
```

В приведенной команде user и password соответствуют имени пользователя и паролю для входа в частный репозиторий. Обратите внимание, если не указать пароль, то его будет предложено ввести при развертывании пакета.

Закончив настройку, можно просто развернуть пакет следующей командой:


```python
poetry publish
```

Если использован частный репозиторий, следует указать его имя Priv с помощью опции -r.

Список команд
Poetry охватывает множество функций, связанных с управлением конфигурацией и публикацией пакетов:

•	shell – при первом запуске этой команды в каталоге вашего проекта Poetry создает виртуальную среду Python, которая будет связана с проектом. Вместо создания папки, содержащей библиотеки зависимостей, Poetry создает среду на глобальном системном пути. Как только эта виртуальная среда будет создана, ее можно активировать в любое время, запустив poetry shell в каталоге проекта.

•	install – устанавливает зависимости, указанные в pyproject.toml. При первой установке создается файл .lock, который содержит фактические номера версий каждого установленного пакета. Номера версий в .lock приоритетнее, чем находящиеся в pyproject.toml.

•	update – если для пакетов в pyproject.toml существуют более новые версии, они будут установлены, и .lock файл будет обновлен.

•	add – добавление зависимости к pyproject.toml. Пакет устанавливается сразу после применения команды.

•	remove – удаление библиотеки зависимостей.

•	config – пути к текущей виртуальной среде или ее переменные. Передача опции --list вернет текущие значения конфигурации.

•	check – проверяет pyproject.toml на наличие ошибок.

•	show – возвращает вывод всех пакетов, установленных в проекте, включая зависимости зависимостей.

•	run – выполняет сценарий, заданный в разделе tool.poetry.scripts в pyproject.toml.

•	build – собирает исходники и пакеты в архив.

•	publish – публикует вывод предыдущей сборки во внешнем репозитории, например, в PyPi.

Чтобы получить справку из командной строки с полным списком команд, просто вызовите poetry.

В экосистеме Python часто появляются различные инструменты, в данном случае рассмотрен инструмет для управления зависимостями. С помощью него появляется возможность чисто и эффективно структурировать свои проекты Python и облегчить управление ими.
